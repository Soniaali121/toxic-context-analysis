```python
import pandas as pd
import re
from IPython.display import display
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', 15)
```

```python
df = pd.read_csv("test.csv")

toxic_keywords = ['bitch', 'fuck', 'stupid', ' idiot', ' hate', ' sicken', 'slap', 'pathetic ' ]
```

```python
def label_toxicity(text):
    if not isinstance(text, str): return 0
    text = text.lower()


    if any(word in text for word in toxic_keywords):
        return 1
    return 0
```

```python
df['is_toxic'] = df['comment_text'].apply(label_toxicity)


identity_groups = {
    'religion':['muslim', 'islam', ' christian', 'jew', ' jewish', 'hindu', 'atheiest'],
    'gender': ['women','woman', 'man', 'gay', 'lesbian', 'trangender', 'gender'],
    'ethnicity': ['black', ' white', 'asian', 'african', 'hispanic', 'latino', 'indian', 'race','ethnic']
}
```

```python
def categorize_identity(text):
    if not isinstance(text,str): return " None"
    text = text.lower()
    for category, keywords in identity_groups.items():
        if any(re.search(r'\b' + word + r'b', text) for word in keywords):
            return category
    return "None"

df['identity_category'] = df['comment_text'].apply(categorize_identity)
analysis_df = df[df['identity_category'] != "None"].copy()
analysis_df.to_csv('labeled_and_categorized.csv', index=False)
print("--- PROFESSOR'S REQUIRED VIEW ---")
analysis_df[['comment_text', 'is_toxic', 'identity_category']].head(5)

```
